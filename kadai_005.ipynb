```python
# Step1.matplotlib.pyplotメソッドとnumpyをインポートするコードを記述する
import numpy as np
import matplotlib.pyplot as plt

# Step2.6個のグラフを表示する領域を作成する
fig, axes = plt.subplots(2, 3, tight_layout=True)

# Step3-1.折れ線グラフ
y0 = [1, 2, -5, 2]
axes[0,0].plot(y0)

# Step3-2.sin関数
x = np.linspace(0, 10, 100)
y = 2 + 2 * np.sin(2 * x)

axes[0,1].plot(x, y, linewidth=2.0)
axes[0,1].set(xlim=(0, 10), xticks=np.arange(0, 10),
       ylim=(-1, 5), yticks=np.arange(-1, 6))

# Step3-3.ヒストグラム
x = np.random .normal(15, 5, 2000)
axes[0,2].hist(x)

# Step3-4.散布図
np.random.seed(3)
x = 4 + np.random.normal(0, 2, 60)
y = 4 + np.random.normal(0, 2, len(x))
axes[1,0].scatter(x, y)

# Step4.グラフを表示させる行「plt.show()」を書く
# plot.show()
```
